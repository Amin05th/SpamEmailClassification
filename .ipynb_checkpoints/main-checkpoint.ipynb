# Data Processing

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler, Normalizer
from sklearn.utils.class_weight import compute_class_weight
```

```python
class FeatureScaler(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        sorted_corr_spam = corr_spam.sort_values(ascending=False)
        ct = ColumnTransformer([
            ("minmaxscaler", MinMaxScaler(), sorted_corr_spam[:7].index),
            ("standardscaler", Normalizer(), sorted_corr_spam[7:].index)
        ])
        X = ct.fit_transform(X)
        return X
```

# Model tuning

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns="spam")
y = df["spam"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
scaler = FeatureScaler()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.ensemble import IsolationForest
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV

trans_X_train = scaler.fit_transform(X_train)
trans_X_test = scaler.fit_transform(X_test)
```

```python
logistic_param_grid = {"solver": ["newton-cg", "lbfgs", "sag", "saga"]}
multi_nb_param_grid = {"alpha": [0.1, 0.3, 0.5], "fit_prior": [True, False]}
bern_nb_param_grid = {"binarize": [0.1, 0.3], "alpha": [0.1, 0.3], "fit_prior": [True, False]}
iso_forest_param_grid = {"n_estimators": [1600, 1800, 2000]}
```

```python
model_list = {
    "LogisticRegression": GridSearchCV(LogisticRegression(penalty="l2", class_weight="balanced", max_iter=10000), param_grid=logistic_param_grid),
    "MultinomialNB": GridSearchCV(MultinomialNB(), param_grid=multi_nb_param_grid),
    "GaussianNB": GaussianNB(),
    "BernoulliNB": GridSearchCV(BernoulliNB(), param_grid=bern_nb_param_grid, scoring="accuracy"),
    "IsolationForest": GridSearchCV(IsolationForest(), param_grid=iso_forest_param_grid, scoring="accuracy")
}
```

# Model evaluation

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, r2_score

def model_evaluation(y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred, average="micro")
    precision = precision_score(y_true, y_pred, average="micro")
    r2 = r2_score(y_true, y_pred)
    
    return accuracy, recall, precision, r2
```

```python
model_score = {}
for model_name, model in model_list.items():
    model.fit(trans_X_train, y_train)
    y_pred = model.predict(trans_X_test)
    model_score[model_name] = model_evaluation(y_test, y_pred)
```

```python
model_scores_df = pd.DataFrame(model_score)
model_scores_df.to_csv("model_scores.csv", index=None)
```

# Final Prediction

```python
predictions = {}
for model_names, model in model_list.items():
    model.fit(trans_X_train, y_train)
    pred = model.predict(trans_X_test)
    predictions[model_names] = pred
```

```python
y_test.shape
```

```python
final_df = pd.DataFrame(index=X_test.index)
final_df['y_true'] = y_test
```

```python
for model, values in predictions.items():
    final_df[model] = values
```

```python
import sqlite3

connection = sqlite3.connect("emails.db")
connection.text_factory = lambda x: str(x, 'utf-8')
c = connection.cursor()

c.execute(
    """
        CREATE TABLE IF NOT EXISTS predictions (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        y_true CHAR,
        LogisticRegression CHAR,
        MultinomialNB CHAR,
        GaussianNB CHAR,
        BernoulliNB CHAR,
        IsolationForest CHAR
        );
    """
)
connection.commit()
```

```python
final_df.to_sql("predictions", con=connection, if_exists="append", index=False)
```
